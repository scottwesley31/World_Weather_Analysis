```python
# Dependencies and Setup
import requests
import gmaps

# Import API key
from config import g_key
```

```python
# Set the parameters to search for a hotel in Paris.
params = {
    "radius": 5000,
    "types": "lodging",
    "key": g_key,
    "location": "48.8566, 2.3522"}

# Use base URL to search for hotels in Paris.
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# Make a request and get the JSON data from the search.
hotels = requests.get(base_url, params=params).json()

hotels
```

```python
# Obtain the length of the "results" dictionary from the hotels request.
# To show more than 20, consult the next_page_token parameter according to the documentation.
len(hotels["results"])
```
